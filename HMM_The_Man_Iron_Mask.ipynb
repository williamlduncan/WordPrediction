# Dictionaries and probabilities of Word Occurrence

The first file has the entire book text. Since it is a long a text, we decided to focus our analysis in the first 2 chapters.

The next two file has the text corpus of Chapter I and Chapter II of the book.

Now we defined some loops to go over each of the files, break the text in lines and create a separate dictionary of the used words in each file.

Thus, we will have the following dictionaries:

* new_dict
* new_dict1
* new_dict2

```python
new_dict ={}
total_count = 0
with open(filename, "r") as fp:   
    for line in fp:     
        words = line.split()
        for word in words:
            word = word.lower()        
        if not word in new_dict:
             new_dict[word] = 1
        else:
            new_dict[word] +=1
            total_count +=1

new_dict1 ={}
total_count = 0
with open(filename_c1, "r") as fp:   
    for line in fp:     
        words = line.split()
        for word in words:
            word = word.lower()        
        if not word in new_dict1:
             new_dict1[word] = 1
        else:
            new_dict1[word] +=1
            total_count +=1

new_dict2 ={}
total_count = 0            
with open(filename_c2, "r") as fp:   
    for line in fp:     
        words = line.split()
        for word in words:
            word = word.lower()        
        if not word in new_dict2:
             new_dict2[word] = 1
        else:
            new_dict2[word] +=1
            total_count +=1
```

# Hidden Markov Model

Hidden Markov Model (HMM) is a statistical Markov model in which the system being modeled is assumed to be a Markov process with unobserved (i.e. hidden) states.

## Mathematical description

A basic hidden Markov model can be described as follows:

$T$ = length of the observation sequence

$N$ = number of states in the model

$M$ = number of observations symbols

$Q$ = ${q_0, q_1, \ldots, q_{N-1} }$ = distinct states of the Markov process

$V$ = ${0, 1, \ldots, M-1 }$ = set of possible observations

$A$ = state transition probabilities

$B$ = observation probability matrix

$\pi$ = initial state distribution

$O$ = $(o_0, O_1, \ldots, o_{T-1})$ = observation sequence

Now let's define our hidden states. We also define our starting probabilities, i.e., the initial state disreibution $$ \pi $$

```python
np.random.seed(20190326)

hidden_states = ['chap1', 'chap2']
pi = [0.5, 0.5]
state_space = pd.Series(pi, index=hidden_states, name='states')
print(state_space)
print('\n', state_space.sum())
```

Now we need to define the state transition probabilities, i.e., the matrix **A** above.

```python
a_df = pd.DataFrame(columns=hidden_states, index=hidden_states)
a_df.loc[hidden_states[0]] = [0.7, 0.3]
a_df.loc[hidden_states[1]] = [0.4, 0.6]
```

Let's check our state transition probabilities matrix **A** :

```python
print(a_df)

a = a_df.values
print('\n', a, a.shape, '\n')
print(a_df.sum(axis=1))
```

As expected the probabilities for each hidden state sum to 1.

Let's update the new_dict1 with the data from new_dict2. This way we will have in one single new_dict1 a list of the words that appears in both dictionaries (our hidden states).

```python
new_dict1.update(new_dict2)
observable_states = new_dict1.keys()
```

Now create another 2 disctionaries populating with the probabilities of word occurence in each hidden state.

```python
words_probabilities_1 ={}
for word in new_dict:
    if not word in words_probabilities1.index.values:
        words_probabilities_1[word] = 0
    else:
        words_probabilities_1[word] = words_probabilities1.loc[word, 0]

words_probabilities_1 = pd.DataFrame.from_dict(words_probabilities_1, orient='index')

words_probabilities_2 ={}
for word in new_dict:
    if not word in words_probabilities2.index.values:
        words_probabilities_2[word] = 0
    else:
        words_probabilities_2[word] = words_probabilities2.loc[word, 0]

words_probabilities_2 = pd.DataFrame.from_dict(words_probabilities_2, orient='index')
```

Having both dictionaries (latter transformed to DataFrames) allow us to define the matrix **B**, i.e., the observation probability matrix.

```python
b_df = pd.DataFrame(columns=observable_states, index=hidden_states)
b_df.loc[hidden_states[0]] = words_probabilities_1.loc[:,0]
b_df.loc[hidden_states[1]] = words_probabilities_2.loc[:,0]
```

```python
print(b_df)
b = b_df.values
print('\n', b, b.shape, '\n')
print(b_df.sum(axis=1))
```

## Multinomial HMM 

Since we have all the information needed, this allows us to  our Hidden Markov Model. We opted to use a Multinomial HMM.

```python
model = hmm.MultinomialHMM(n_components=2)
model.startprob_ = np.array(pi)
model.transmat_ = a_df.astype(float)
model.emissionprob_ = b_df.astype(float) 
```
